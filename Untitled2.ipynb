# График отображающий соотношение скилла(признак Overall) футболиста и его зарплаты


```python
import pandas as pd
import numpy as no
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore')
%matplotlib inline
```

```python
df = pd.read_csv('data/CompleteDataset.csv')
```

Конвертируем тип  признака Wage  в числовой

```python
def convert_money(x):
    if x[-1].isdigit():
        return float(x[1:])
    elif x[-1].isalpha():
        return float(x[1:-1]) * 1000000 if x[-1]=='M' else float(x[1:-1]) * 1000
    else:
        return None
df['Wage'] = df['Wage'].apply(convert_money)
```

```python
plt.plot('Overall','Wage',data=df,linestyle=None,marker='o');
plt.title('Соотношение зарплаты футболиста и его скилла')
plt.xlabel('Скилл')
plt.ylabel('Зарплата в евро')
plt.yticks(range(0,600000,50000));
```
