# Графики  отражающие особенности распределения возраста футболистов

```python
import pandas as pd
import numpy as no
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore')
%matplotlib inline
```

```python
df = pd.read_csv('CompleteDataset.csv')
```

## Гистограмма 

```python
df['Age'].plot.hist(bins=100)
plt.title('Гистограмма распределения возрастов футболистов',fontdict={'fontsize':20,'color':'green'})
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.xticks(range(15,50,2));
```

## Boxplot(ящик с усами)

```python
plt.boxplot(df['Age']);
plt.title('Ящик с усами для признака- Возраст футболиста',fontdict={'fontsize':18,'color':'brown'})
plt.ylabel('Возраст')
plt.yticks(range(15,50,2));
```

Построив 2 графика, мы уже можем увидеть, что медианный возраст футболистов составляет 25 лет.
А основная масса футболистов имеет возраст в диапазоне от 21 до 28 лет.
