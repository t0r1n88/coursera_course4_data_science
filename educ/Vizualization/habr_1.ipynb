### Seaborn
Теперь давайте перейдем к библиотеке seaborn. Seaborn — это по сути более высокоуровневое API на базе библиотеки matplotlib. Seaborn содержит более адекватные дефолтные настройки оформления графиков. Также в библиотеке есть достаточно сложные типы визуализации, которые в matplotlib потребовали бы большого количество кода.

Познакомимся с первым таким "сложным" типом графиков pair plot (scatter plot matrix). Эта визуализация поможет нам посмотреть на одной картинке, как связаны между собой различные признаки.

```python
# с svg pairplot браузер начинает тормозить
%config InlineBackend.figure_format='png'
sns.pairplot(df[['Global_Sales', 'Critic_Score', 'Critic_Count', 
                 'User_Score', 'User_Count']]);
```

```python
#Также с помощью seaborn можно построить распределение, для примера посмотрим на распределение оценок критиков Critic_Score.
#Для этого построим distplot. По default'у на графике отображается гистограмма и kernel 
%config InlineBackend.figure_format='svg'
sns.distplot(df['Critic_Score']);
```

Для того чтобы подробнее посмотреть на взаимосвязь двух численных признаков, есть еще и joint_plot – это гибрид scatter plot и histogram (отображаются также гистограммы распределений признаков). Еще один полезный тип графиков – это box plot. Давайте сравним пользовательские оценки игр для топ-5 крупнейших игровых платформ.


```python
# Считаем количество игр для каждой платформы
# df['Platform'].value_counts()
# Сортируем по убыванию и выбираем 5 самых популярных
# df['Platform'].value_counts().sort_values(ascending=False).head(5)
# Получаем значения индекса этих 5 значений
#df['Platform'].value_counts().sort_values(ascending=False).head(5).index.values
top_platforms=df['Platform'].value_counts().sort_values(ascending=False).head(5).index.values
top_platforms
sns.boxplot(x='Critic_Score',y='Platform',data=df[df['Platform'].isin(top_platforms)],orient='h');
```

Думаю, стоит обсудить немного подробнее, как же понимать box plot. Box plot состоит из коробки (поэтому он и называется box plot), усиков и точек. Коробка показывает интерквантильный размах распределения, то есть соответственно 25% (Q1) и 75% (Q3) процентили. Черта внутри коробки обозначает медиану распределения. С коробкой разобрались, перейдем к усам. Усы отображают весь разброс точек кроме выбросов, то есть минимальные и максимальные значения, которые попадают в промежуток (Q1 - 1.5*IQR, Q3 + 1.5*IQR), где IQR = Q3 - Q1 - интерквантильный размах. Точками на графике обозначаются выбросы (outliers) - те значения, которые не вписываются в промежуток значений, заданный усами графика.

И еще один тип графиков (последний из тех, которые мы рассмотрим в этой статье) - это heat map. Heat map позволяет посмотреть на распределение какого-то численного признака по двум категориальным. Визуализируем суммарные продажи игр по жанрам и игровым платформам.

```python

platform_genre_sales = df.pivot_table(index='Platform',columns='Genre',values='Global_Sales',
                                     aggfunc=sum).fillna(0).applymap(float)
```

```python
sns.heatmap(platform_genre_sales,annot=True,fmt=".1f",linewidths=.5);
```

# Plotly
Мы рассмотрели визуализации на базе библиотеки matplotlib. Однако, это не единственная опция для построения графиков на языке python. Познакомимся также с библиотекой plotly. Plotly - это open-source библиотека, которая позволяет строить интерактивные графики в jupyter.notebook'e без необходимости зарываться в javascript код.

Прелесть интерактивных графиков заключается в том, что можно посмотреть точное численное значение при наведении мыши, скрыть неинтересные ряды в визуализации, приблизить определенный участок графика и т.д.

Перед началом работы импортируем все необходимые модули и инициализируем plotly с помощью команды init_notebook_mode.

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

Для начала построим ***line plot*** с **динамикой** числа вышедших игр и их продаж по годам.

```python
# Группируем по году и суммируем все данные по столбцу Global sales
#df.groupby('Year_of_Release')[['Global_Sales']].sum()
# Обьединяем с другим датафреймом сгруппированым по годам где подсчитано количество игр в каждом году
# подсчет просто через столбец Name и count
# years_df=df.groupby('Year_of_Release')[['Global_Sales']].sum().join(
#     df.groupby('Year_of_Release')[['Name']].count())
# Меняем название колонок
# years_df.columns=['Global_Sales','Number_of_Games']
# Следует запомнить что когда мы производим группировку то индексом
# становятся значения столбца по которому мы производим группировку 
years_df=df.groupby('Year_of_Release')[['Global_Sales']].sum().join(
    df.groupby('Year_of_Release')[['Name']].count())
years_df.columns=['Global_Sales','Number_of_Games']
years_df
```

В plotly строится визуализация объекта Figure, который состоит из данных (массив линий, которые в библиотеке называются traces) и оформления/стиля, за который отвечает объект layout. В простых случаях можно вызывать функцию iplot и просто от массива traces.

```python
years_df.index
```

```python
# Строим объект Figure
# Определяем первую линию
trace0 = go.Scatter(x=years_df.index,y=years_df['Global_Sales'],name='Global Sales')
# По оси абсцисс(т.е оси х) у нас будут года взятые из индекса датафрейма
# По оси ординат(т.е оси у) у нас будет объем продаж
# Определяем вторую линию
trace1 = go.Scatter(x=years_df.index,y=years_df['Number_of_Games'],name='Number of games released')
# По оси абсцисс(т.е оси х) у нас будут года взятые из индекса датафрейма
# По оси ординат(т.е оси у) у нас будет число изданных игр
# Создаем массив линий
data = [trace0,trace1]
# Задаем словарь который будет использовать обьект layer
layer= {'title':'Statistics of video games'}
# Создаем обьект Figure который и будет рисовать наш график
fig = go.Figure(data=data,layout=layer)
# Рисуем диаграмму
# show_link отвечает за показ ссылки на сайт где можно редактировать этот график
iplot(fig,show_link=True)

```

Также можно сразу сохранить график в виде html-файла


```python
plotly.offline.plot(fig,filename='years_stats.html',show_link=False)
```

Посмотрим также на рыночную долю игровых платформ, расчитанную по количеству выпущенных игр и по суммарной выручке. Для этого построим bar chart.

```python
# Снова соединяем 2 группы по такому же принципу
platforms_df = df.groupby('Platform')[['Global_Sales']].sum().join(
df.groupby('Platform')[['Name']].count())
platforms_df.columns = ['Global_Sales', 'Number_of_Games']
# Сортируем по убыванию датафрейм по столбцу Global_sales
# С помощью параметра inplace=True, изменяем сам исходный датафрейм а не копию
platforms_df.sort_values('Global_Sales', ascending=False, inplace=True)
platforms_df
```

```python
# Создаем 1 линию
trace0 = go.Bar(x=platforms_df.index,y=platforms_df['Global_Sales'],name='Global sales')
# создаем 2 линию
trace1 = go.Bar(x=platforms_df.index,y=platforms_df['Number_of_Games'],name='Number of games released')
# Создаем массив линий
data = [trace0,trace1]
layout = {'title':'Share of platforms'}
# Создаем обьект Figure
fig1 = go.Figure(data=data,layout=layout)
iplot(fig1,show_link=False)
```

```python
# В plotly можно построить и box plot. Рассмотрим различия оценок критиков в зависимости от жанра игры.
data=[]
for genre in df['Genre'].unique():
    #Добавляем в массив линий обьекты Box
    data.append(
    go.Box(y=df[df['Genre']==genre]['Critic_Score'],name=genre))
iplot(data,show_link=False)

```
