```python
import pandas as pd
import sqlite3
import pandas_profiling
```

read_sql не устанавливает первичный ключ (id) в качестве индекса. Можно это сделать вручную, добавив аргумент index_col к read_sql.

Если вы много использовали read_csv, вы могли заметить, что у него также есть аргумент index_col. И ведёт он себя точно так же.
Если вы хотите, чтобы dataframe был индексирован несколькими столбцами, в index_col можно указать их список:
df = pd.read_sql("SELECT * from weather_2012 LIMIT 3", con, 
                 index_col=['id', 'date_time'])

```python
# Создаем объект для открытия файла базы данных
con = sqlite3.connect('../../data/weather_2012.sqlite')
# Считываем в датафрейм

# df = pd.read_sql('SELECT * FROM weather_2012 LIMIT 3',con)

# Вариант с установкой идекса сразу
df = pd.read_sql('SELECT * FROM weather_2012',con,index_col='id') 
```

```python
import numpy as np
```

```python
cf.set_config_file(theme='ggplot',sharing='public',offline=True)
```

```python

# Cufflinks can generate random data for different shapes
# Let's generate a single line with 1000 points
cf.datagen.lines(1,1000).iplot()
```

```python
# Generating 4 timeseries 
df=cf.datagen.lines(4,1000)
df.iplot()
```

```python
df3 = pd.DataFrame(np.random.randn(1000, 2), columns=['B', 'C']).cumsum()
df3['A'] = pd.Series(list(range(len(df3))))
df3.iplot(x='A', y='B')
```
